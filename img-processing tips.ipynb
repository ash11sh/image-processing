```
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
```

```
img1 = mpimg.imread('gatsby.jpg')
img2 = mpimg.imread('blade_runner.jpg')
```

## Show multiple images at a time
### method 1

```
(nrows, ncols) = (1,2)

plt.subplot(nrows, ncols, 2) # rows, cols, and index
plt.imshow(img1)
plt.subplot(nrows, ncols, 1)
plt.imshow(img2)
plt.subplots_adjust(left=0.075, bottom=0.05, right=0.95, top=0.95, wspace=0.45, hspace=0.45)
plt.show()
```

### method 2

```
nrows,ncols = 1,2
fig, axs = plt.subplots(nrows,ncols, figsize=(10,10))
axs[0].imshow(img1)
axs[0].set_title('gatsby')
axs[0].set_xticks([]), axs[0].set_yticks([])
axs[1].imshow(img2)
axs[1].set_title('blade runnr')
axs[1].set_xticks([]), axs[1].set_yticks([])
plt.show()
```

## Draw rectangle on image

```
x1,y1,x2,y2 = 200,200,550,657
rect_img = cv2.rectangle(img2.copy(),(x1,y1),(x2,y2),(255),-1)
```

```
plt.imshow(rect_img)
```

## Draw a rectangle with text on it

```
fig, axs = plt.subplots(figsize=(10,10))
axs.imshow(img2)

x1,y1,w,h = 180,200,460,550
axs.add_patch(plt.Rectangle((x1, y1), w, h, fill=False, edgecolor='red', linewidth=3.5))
axs.text(x1,y1-30,'HOLOGRAM',bbox=dict(facecolor='yellow', alpha=0.7), fontsize=20, color='white')
plt.show()
```
